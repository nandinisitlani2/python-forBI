```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("CreditCard.csv")
```

```python
df.head()
```

```python
df=pd.get_dummies(data=df,
                  drop_first=True)
```

```python
y=df["card_yes"]
x=df.drop(columns="card_yes")
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,
                                               random_state=1502,
                                               stratify=y)

## stratify y is used to get a balanced split based on y because our data is imbalanced
```

```python
from sklearn.ensemble import RandomForestClassifier

model=RandomForestClassifier(n_estimators=300,
                             random_state=1502)
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import classification_report,f1_score
```

```python
print(classification_report(y_test,y_pred))
```

## Feature Importance

```python
importance=pd.Series(model.feature_importances_,
                     index=x_train.columns.values)
importance.nlargest(5).plot(kind="barh",
                            figsize=(10,6))
```

```python
from sklearn.model_selection import ParameterGrid

param_grid={"n_estimators":range(50,500,50)}
grid=ParameterGrid(param_grid)
grid[1]
```

```python
f1score=[]

for params in grid:
  model=RandomForestClassifier(n_estimators=params["n_estimators"],
                               random_state=1502)
  model.fit(x_train,y_train)
  predictions=model.predict(x_test)
  f1=f1_score(y_test,predictions)
  f1score.append(f1)
```

```python
best_param=grid[np.argmax(f1)]
print(best_param)
```
